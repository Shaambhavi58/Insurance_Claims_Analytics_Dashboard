# Insurance Claims Analytics – EDA & KPIs

Quick EDA on synthetic claims data. Run the ETL first to create `data/claims.db`.

```python
import pandas as pd
import matplotlib.pyplot as plt
claims = pd.read_csv('../data/claims_sample.csv')
claims.head()
```

```python
claims['claim_date'] = pd.to_datetime(claims['claim_date'])
claims.groupby('product')['claim_amount'].sum().plot(kind='bar')
plt.title('Total Claim Amount by Product')
plt.xlabel('Product')
plt.ylabel('Amount')
plt.tight_layout()
plt.show()
```
